# QUEUE

## What is queue?

### we can define queue as an ordered list , linear abstract data structure which follows FIFO first in first out principal where we can add element ( enqueue operation) from the rare end of the queue and delete element (dequeue) from the front end of the queue 

## Operations: 
### Enqueue(): to insert the data in queue from the rare end 
### Dequeue(): to delete the data from queue from the front end
### front(): it is similar to peek operation  in stack wherein it tends to display the data from front end 
### isFull(): it check whether the queue is full or not 
### isEmpty(): it check whether the queue is empty or not 

## overflow condition :
### it arises when the queue is completely full and it still we are trying to add data to the queue 

## underflow condition: 
### it arises when queue is empty 

# Drawback of Queue 
### condition where rear==front and before that empty space is vacant and still if we try to add new data it is raising exception stating queue is full (memory wastage ) it can be over come using circular queue 

# Application of queue 
### 1. customer care call center 
### 2. in printing machine when it is shared with 4-5 PCs 

## when we try to implement queue using array 
### 1. chance of overflow is high 
### 2. time complexity is of O(n) for copy operation in case queue get full and all data has to copy to new biger array 
### 3. if we take to large array and cannot able to utilize the space fully then leads to wastage of memory 

## implementing using linked list 
### for enqueue time complexity O(1)
### for dequeue time complexity O(n) 
### in order to over come time complexity of O(n) for dequeue operation we need to create one more pointer at the rear end of the queue 




```python
# defining node class 
class node:
    # initialization of variables 
    def __init__(self,value):
        self.data=value
        self.next=None # initially it is assigned to none 
# defining queue class 
class Queue :
    # initialization of front are rear pointer 
    def __init__(self):
        self.rear = None
        self.front = None 
    # defining __str__ special method to print result as per req    
    def __str__(self):
        t = self.front
        all_element = "\n [queue front]"
        num_element = 1 
        while t: 
            all_element=all_element + " [#" + str(num_element) + "] " + str(t.data) + " <--"
            t=t.next
            num_element+=1
        return all_element + "[queue rear] \n"
    
    # defining method to check whether queue is empty or not 
    def isEmpty(self):
        return self.front==None
    
    # defining insertion method 
    def Enqueue(self,value):
        #creating instance of node class 
        newnode=node(value)
        
        # checking if queue is empty or not if so then both pointer should point to the newly added node
        if self.rear==None:
            self.front=newnode
            self.rear = newnode
            return
        # if queue is not empty in that case we have to pass the addrress of new node to prev node and shift the
        # rear pointer to newnode 
        self.rear.next=newnode
        self.rear=newnode
        
    # defining deletion method 
    def Dequeue (self):
        #checking whether the queue is empty or not 
        if self.isEmpty():
            return
        # if not empty in that case store the address of front node into temp variable and then shift the front 
        # pointer to the second node from the top which means top node link get break and is eligible for garbage
        # coollection 
        t=self.front
        self.front.next=t.next
        
        if self.front == None:
            self.rear == None
def main():
    q=Queue()
    q.Enqueue("person1")
    q.Enqueue("person2")
    q.Enqueue("person3")
    q.Enqueue("person4")
        
    print(q)
    
    q.Dequeue()
    q.Dequeue()
    q.Enqueue("person5")
    q.Enqueue("person6")
    
    print(q)
if __name__=="__main__":
    main()
        
```
