## Implementation of linked list ,append operation (beginning , in between and at last) , counting the number of nodes , traverse operation, delete operation ( beginning, end , delete by value )

##### 1.  to implement linked list we have to define head which has to hold the address of the nodes which we will be defining,
##### 2. we have to define node class and inside node we have data variable which will hold the value and next which will hold the address of next    node 
##### 3. we have to create the object of the linked list class and node class 
##### 4. after creating the object we have to link the head with nodes and each time it iterate it has to print the values of respective node untill it        encounter null 


```python
# defining node class 
class node: 
    
    # defining constuctor for the node class in order to initialize the variables 
    
    def __init__(self,value):
        self.data=value
        self.next=None

#defining Linkedlist class 
class Linkedlist:
    
    # defining constructor to initialize varaible of linked list class 
    def __init__(self):
        self.head=None
        
    
    # defining append method in order to append new element in the linked list 
    
    
    # it involve following steps 
    # 1. we have to create a method as shown below 
    # 2. we are creating new node using newnode=node(value)
    # 3. now we have to add address of self.head to newnode
    # 4. finaly we have to move our cursor to newnode instead of self.head 
    
    
    def appendingBeginning(self,new_value):
        print("appending at the beginning")
        NewNode=node(new_value) # object of node is being created 
        NewNode.next = self.head # by doing so we are basically assigning address of self.head to NewNode
        self.head = NewNode # in this step we are shifting our pointer from self.head to newnode bascialy we are making
                            # NewNode as head and we want our cusor to point to this node so we are doing the folloeing step
    
    
    # defining append method to append new value at the last of the linked list 
    # it involves following steps 
    # 1. when the linked list is empty inside append method we have to create the instance of node class and make the 
    #    node as head for the linked list 
    # 2. if not then we have to traverse till the end and there we have to assign the address of new node created to the 
    #    previous node 
    
    def appendLast(self,new_value):
        print("appending at the last")
        NewNode=node(new_value) # creating instance of node class 
        # checking if linked list is empty or not if it is empty then in such case we have to assign self.head to address 
        # of new node --> NewNode=node()
        if self.head==None:
            self.head=NewNode
            return
        
        # if linked list is not empty in such case we have to traverse till the end of the linked list and at the last when
        # we encounter last.next=None there we have to assign last.next = NewNode address of object of node class created 
        last=self.head
        while(last.next):
            last=last.next
        last.next=NewNode
    
    # defining a method to insert new node after a given node 
    
    # here we have to check whether the node after which we want to add new node is vaild or not in simple words it 
    # exist within linkedlist or not 
    # we have to create instance of the new node 
    # after this new node will hold the address which was being hold by prev node 
    # after performing above step the link between prev node and next node get cut and we have to set new link between 
    # new node prev node and next node 
    # prev node will be assigned address of new node created 
    
    def appendingAfter(self,prevNode,new_value):
        print("appending after given node ")
        # checking whether the given node is within the linked list or not 
        if prevNode is None:
            print("please enter a valid node")
            return
        # creating instance of node 
        NewNode=node(new_value)
        
        # breaking the old link and relinking 
        
        NewNode.next=prevNode.next
        prevNode.next=NewNode
    # defining the delete method    
    def delete_start(self):
        print("deleting node from begining of the linked list  ")
        # checking if tthe linkrd list is empty or not if it is empty that means we cannot delete the node 
        if self.head ==None:
            print("linked list is empty we cannot perform deletetion operation ")
            return
        # if the linked list is not empty in that case we have to store the address of 1st node in a variable
        # from there we have to point to the next node to fetch the address of the next node 
        # once we have address of next node we will move the pointer to the next node by assigning the address of 
        # next node to the self.head variable 
        t=self.head
        self.head=t.next
        t.next=None
        
        
    # defining deleting operation to delete node at end 
    def delete_end(self):
        # checking whether the given linked list is empty or not 
        if self.head==None:
            print("linked list is empty")
        # checking whether the linked list has only one node or not 
        if self.head.next==None:
            print("it has only one node ")
            self.head=None
        # here using while loop we are traversing through the linked list untill from the second last node we get the 
        # last node 
        n = self.head
        while(n.next.next is not None):
            n=n.next
        #once we get the second last node from there we have to reassign the address of next node to None in order to drop 
        # the node coming after that node 
        n.next=None
        
        
    def deleteByValue(self,value):
        # checking for the linked list is empty or not 
        if self.head==None:
            print("we cannot perform deletion operation as the linked list is empty!")
            return
        # checking whether the data present in the 1st node matches the value passed by user
        # if condition satisfied in that case we have to move the pointer to next node 
        if value==self.head.data:
            print("we are in the first node and now we have to shift pointer to next node ")
            t=self.head
            self.head=t.next
            return
        n = self.head
        while(n.next is not None):
            # checking for the previous node with respect to the node passed by user if we get break the loop there only 
            if value==n.next.data:
                break
            # if we didnt find the prev node wrt given node by user we have to reassign the next address to the counter 
            # variable again and again till we get prev node 
            n=n.next
            
        # we arrive here because of 2 reasons either we get the prev node or condition fails n.next == None 
        # if n.next==none then simply pass a msg stating node not found 
        if n.next==None:
            print("node is not present in the linked list")
            
        # once we get prev node just reassign the address present in the prev node to the address of node which comes after
        # node passed by the user 
        else:
            # here we have to skip the node to which the value is matching and we have to assign the reference of 
            # node coming after the node whose value matches the value given by user 
            n.next=n.next.next
            
            
            
        
    # defing method to print the values present in the node of linked list 
    def printingData(self):
        print("---printing the element of linked list--- ")
        print()
        
        # to print the values of the node we have used while loop 
        # initialize h1 to a certain address followed by checking conditon whether the address evaluate to True or not 
        # followed by printing the values of the node and lastly updating the vallues of h1 in order to achieve stopping
        # condition 
        h1=self.head
        while(h1):
            print(h1.data)
            h1=h1.next
               
            
    # defining a method to count total numbers of nodes available in the linked list 
    
    def countingNodes(self):
        countNode=1
        element=self.head
        while(element):
            countNode +=1
            element=element.next
        print()
        print("---"*30)
        print(f"number of nodes available in the linked list : {countNode}")
        
  
        
# defining main class                     
def main():
    ll1=Linkedlist() # object creation for linkedlist class 
    n1=node(20)      # object creation of node class 
    n2=node(30)      #object creation of node class
    n3=node(40)      #object creation of node class
    
    # linking node with head because in linkedlist cursor will initially point to head and from head to nodes 
    
    ll1.head=n1 # head is having address of n1 
    n1.next=n2  # n1 is having address of n2 
    n2.next=n3  # n2 is having addresss of n3
    n3.next=None
    
    
    # printing before appending at the begining 
    
    ll1.printingData()
    
    # calling appendingBegning()
    
    ll1.appendingBeginning(10)
    
    # printing before appending at the last 
    
    ll1.printingData()
    
    # calling appendlast()
    
    ll1.appendLast(50)
    
    # printing before adding element at the given node 
    
    ll1.printingData()
    
    # calling appendAfter()
    
    ll1.appendingAfter(n2,35)
    
    # calling printingData() using object reference in order to execute the printingData() method 
    
    ll1.printingData()
    
    # deleting the node form the begning of the linked list 
    ll1.delete_start()
    
    ll1.printingData()
    
    ll1.delete_end()
    ll1.printingData()
    
    ll1.deleteByValue(90)
    ll1.printingData()
    
    # calling countnode method using object refrence 
    ll1.countingNodes()
    
    
    
if __name__=="__main__":
    main()
        
        
        
        
        
        
        
```
