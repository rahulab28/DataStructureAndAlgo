#                                                     STACK


## what is stack ? 

### A linear data structure wherein we can add or remove the elements from one end i.e top of the stack and it follow FILO (first in last out principal or last in first out LIFO)

## NOTE:  stack can store homogenous data that means all the data present inside the stack is of same type either it can be integer, float etc 

## operations :

### push(): it is used to push or insert the element from the top of the stack or we can say at the beginning of the node of the stack 

### pop(): it is used to pop out the very last element inserted to the stack it basically first return the element and then delete the element from the stack 

### peek(): it is used to return or display the last element addded to the stack or the element present in the top of the stack 

### isEmpty() : it return True if the stack is empty else return False 

### isFull() : it is used to check whether the stack is full or not if yes it returns True else return False 

## time complexity of above mention methods is of O(1) 

## memory allocation for stack :

### 1. static memory allocation using array

### 2. dynamic memeory allocation using linked list ( since we are using linked list we will not face problem of overflow of memory ) 

## memory overflow : (condition occures rarely because we are implementing stack using linked list) 

### it is the sitution wherein there is no space available for insertion of new element still we try to add new element the this condition arises and it can be handled  by rasing exception stating stack is completely full no extra data can be added !! 

## memory underflow: 

### it is the sitution wherein there is no data available at the stack and in that case we can handle the sitution by rasing exception stating stack is empty !! 

## Application of stack 

### 1. in reversing the string 
### 2. undo mechanism in the word 
### 3. recursion function 
### 4. checking the balance of paranthesis in the prog lang 
### 5. infix to post/prefix expression
